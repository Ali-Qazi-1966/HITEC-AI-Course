```python
# Lab 08 - Task 01
# Alpha-Beta Pruning Algorithm

print("==== Alpha-Beta Pruning ====\n")

import math

def alpha_beta(depth, node_index, is_max, scores, alpha, beta, max_depth):
    # Base case
    if depth == max_depth:
        return scores[node_index]

    if is_max:
        best = -math.inf
        for i in range(2):
            value = alpha_beta(depth + 1, node_index * 2 + i,
                               False, scores, alpha, beta, max_depth)
            best = max(best, value)
            alpha = max(alpha, best)

            if beta <= alpha:
                break   # pruning
        return best

    else:
        best = math.inf
        for i in range(2):
            value = alpha_beta(depth + 1, node_index * 2 + i,
                               True, scores, alpha, beta, max_depth)
            best = min(best, value)
            beta = min(beta, best)

            if beta <= alpha:
                break   # pruning
        return best


# Leaf node values (H, I, J, K, L, M, N, O)
scores = [4, 2, -3, -6, 7, 0, 5, 8]

depth = int(math.log2(len(scores)))

result = alpha_beta(0, 0, True, scores, -math.inf, math.inf, depth)

print("Optimal value using Alpha-Beta Pruning:", result)
```

```python
# Lab 08 - Task 02
# Minimax Algorithm for Tic-Tac-Toe

print("\n==== Tic-Tac-Toe using Minimax ====\n")

# Function to check winner
def check_winner(board):
    win_states = [
        [0,1,2], [3,4,5], [6,7,8],   # rows
        [0,3,6], [1,4,7], [2,5,8],   # columns
        [0,4,8], [2,4,6]             # diagonals
    ]

    for state in win_states:
        if board[state[0]] == board[state[1]] == board[state[2]] != ' ':
            return board[state[0]]
    return None

# Minimax function
def minimax(board, is_max):
    winner = check_winner(board)

    if winner == 'X':
        return 1
    elif winner == 'O':
        return -1
    elif ' ' not in board:
        return 0

    if is_max:
        best = -100
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'
                best = max(best, minimax(board, False))
                board[i] = ' '
        return best
    else:
        best = 100
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'
                best = min(best, minimax(board, True))
                board[i] = ' '
        return best


# Example board state
board = ['X', 'O', 'X',
         ' ', 'O', ' ',
         ' ', ' ', 'X']

print("Board State:")
print(board)

value = minimax(board, True)

print("\nOptimal value for X:", value)
```
