```python
# Lab 4 - Task 01
# Employee Management System using Inheritance

print("==== Employee Management System ====\n")

# Base class 1
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Base class 2
class Job:
    def __init__(self, designation, salary):
        self.designation = designation
        self.salary = salary

# Derived class (inherits from both Person and Job)
class Employee(Person, Job):
    def __init__(self, name, age, designation, salary):
        # Initialize both parent classes
        Person.__init__(self, name, age)
        Job.__init__(self, designation, salary)

    def display_details(self):
        print("Employee Details:")
        print("=================")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Designation:", self.designation)
        print("Salary:", self.salary)

# Taking input from user
name = input("Enter employee name: ")
age = int(input("Enter employee age: "))
designation = input("Enter employee designation: ")
salary = float(input("Enter employee salary: "))

# Creating object of Employee
emp = Employee(name, age, designation, salary)

print("\n==== Displaying Employee Information ====\n")
emp.display_details()
```

```python
# Lab 4 - Task 02
# Write a Python class named Circle constructed by a radius
# and two methods to compute area and perimeter.

print("==== Circle Calculator ====\n")

import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

    def perimeter(self):
        return 2 * math.pi * self.radius

# Take input from user
r = float(input("Enter radius of the circle: "))

# Create Circle object
circle = Circle(r)

print("\n==== Circle Details ====\n")
print("Radius:", r)
print("Area:", circle.area())
print("Perimeter:", circle.perimeter())
```

```python
# Lab 4 - Task 03
# Write a Python class to implement pow(x, n) without using math.pow().
# Handle both positive and negative powers.

print("==== Power Function (pow(base, exponent)) ====\n")

class PowerCalculator:
    def calculate_power(self, base, exponent):
        power_result = 1

        if exponent > 0:
            for i in range(exponent):
                power_result = power_result * base
        elif exponent < 0:
            for i in range(-exponent):
                power_result = power_result * base
            power_result = 1 / power_result
        else:
            power_result = 1

        return power_result

# Take input from user
base = float(input("Enter the base number: "))
exponent = int(input("Enter the exponent: "))

# Create object of PowerCalculator
calculator = PowerCalculator()

# Display result
print("\nResult of", base, "^", exponent, "=", calculator.calculate_power(base, exponent))
```

```python
# Lab 4 - Task 04 (Inheritance Example)
# Write a Python program based on the given class diagram.

print("==== Shopping System ====\n")

# Base Class
class Item:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price

    def viewFullDescription(self):
        print("\nItem Name:", self.name)
        print("Description:", self.description)
        print("Price:", self.price)

    def addToShoppingBasket(self):
        print(f"{self.name} has been added to your shopping basket.")

    def removeFromShoppingBasket(self):
        print(f"{self.name} has been removed from your shopping basket.")


# Derived Class 1
class MP3(Item):
    def __init__(self, name, description, price, artist, duration):
        super().__init__(name, description, price)
        self.artist = artist
        self.duration = duration

    def play(self):
        print(f"Now playing '{self.name}' by {self.artist} ({self.duration} mins).")

    def download(self):
        print(f"Downloading '{self.name}'...")


# Derived Class 2
class DVD(Item):
    def __init__(self, name, description, price, certificate, duration, actors):
        super().__init__(name, description, price)
        self.certificate = certificate
        self.duration = duration
        self.actors = actors

    def viewTrailer(self):
        print(f"Playing trailer for '{self.name}' featuring {self.actors}.")


# Derived Class 3
class Book(Item):
    def __init__(self, name, description, price, author, numberOfPages, genre):
        super().__init__(name, description, price)
        self.author = author
        self.numberOfPages = numberOfPages
        self.genre = genre

    def previewContent(self):
        print(f"Previewing first few pages of '{self.name}' by {self.author}.")


# Example usage
mp3_item = MP3("Song", "music track", 150, "Ali Abbas Qazi", 3.5)
dvd_item = DVD("Movie", "Action movie", 800, "PG-13", 120, "Ali Abbas Qazi")
book_item = Book("Novel", "Fiction novel", 500, "Ali Abbas Qazi", 320, "Fiction")

mp3_item.viewFullDescription()
mp3_item.play()
mp3_item.download()
mp3_item.addToShoppingBasket()

dvd_item.viewFullDescription()
dvd_item.viewTrailer()

book_item.viewFullDescription()
book_item.previewContent()
```
