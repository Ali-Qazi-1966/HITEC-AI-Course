```python
# Lab 7 - Task 01
# Apply BFS and DFS on Graph Problem 1

print("==== Graph Problem 1 - BFS & DFS ====\n")

from collections import deque

# Graph based on the picture
graph1 = {
    0: [1, 2],
    1: [3, 4, 5],
    2: [5, 6],
    3: [],
    4: [],
    5: [],
    6: []
}

# Breadth First Search
def bfs(start):
    print("BFS Traversal:\n")
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            for neighbor in graph1[node]:
                queue.append(neighbor)
    print("\n")

# Depth First Search
def dfs(start):
    print("DFS Traversal:\n")
    visited = set()
    stack = [start]

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            for neighbor in reversed(graph1[node]):
                stack.append(neighbor)
    print("\n")

# Run both
bfs(0)
dfs(0)
```
